### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link



```python
#passo a passo do seu projeto
#passo1: importar a basa de dados
import pandas as pd
df= pd.read_csv("cancelamentos.csv")
```

```python
#passo2: visualizar a base de dados
display(df)
```

```python
#passo3: tratamento de erros
display(df.info())
```

```python
#tratar valores vazios
df= df.dropna()
display(df.info())
```

```python
#tratar informacoes desnecessarias
#axis: 1 (coluna)/0 (linha)
df= df.drop("CustomerID", axis=1)
df
```

```python
#passo4: Analise inicial dos dados (entender os cancelamentos)
df["cancelou"].value_counts()
df["cancelou"].value_counts(normalize=True)
```

```python
df["duracao_contrato"].value_counts()
df["duracao_contrato"].value_counts(normalize=True).map("{:.1%}".format)
```

```python
df.groupby("duracao_contrato").count()
```

```python
df.groupby("duracao_contrato").sum()
```

```python
df.groupby("duracao_contrato").mean()
```
